```python
import pandas as pd
import numpy as np
import glob
```

```python
# 取得所有 CSV 檔案的路徑
csv_files = glob.glob('高雄市資料/*.csv')

df = pd.DataFrame()

for file in csv_files:
    data = pd.read_csv(file, header=0,skiprows=[1])
    name = str(file).split('\\')[1].split('.')[0].split('-')
    data['測站'] = name[0]
    data['年份'] = name[1]
    data['月份'] = name[2]
    print(file)
    df = pd.concat([df, data], axis=0,ignore_index=True);df
df.to_csv('高雄市原始資料.csv', index=False, encoding='utf-8-sig')
```

```python
df_na = df.replace('...', float('nan'))
df_na = df_na.replace('/', float('nan'))
df_na = df_na.replace('X', float('nan'))
df_na = df_na.replace('T', float('nan'))
df_na = df_na.replace('&', float('nan'));df_na
df_na.to_csv('基隆市.csv', index=False, encoding='utf-8-sig')
```
